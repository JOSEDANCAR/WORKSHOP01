Insertar el Dataset en la tabla creada en MySQL Workbench ** No Ejecutar Mas **

```python
import mysql.connector
import pandas as pd

# Configurar los detalles de la conexión
config = {
    'user': 'root',
    'password': 'root',
    'host': 'localhost',
    'database': 'candidates'
}

# Conectar a la base de datos
db_connection = mysql.connector.connect(**config)
cursor = db_connection.cursor()

# Ruta al archivo CSV
csv_file_path = r'C:\Users\josed\OneDrive\Documentos\UNIVERSIDAD\ETL\WORKSHOP 1 - JOSE CARRERA\candidates.csv'

# Leer el archivo CSV en un DataFrame de pandas
candidates_data = pd.read_csv(csv_file_path, delimiter=';')

# Verificar los nombres de las columnas
print(candidates_data.columns)

# Asegurarse de usar los nombres correctos de las columnas
candidates_data['Application Date'] = pd.to_datetime(candidates_data['Application Date'], errors='coerce').dt.strftime('%Y-%m-%d')
candidates_data['YOE'] = pd.to_numeric(candidates_data['YOE'], errors='coerce').fillna(0).astype(int)

# Insertar los datos del CSV en la tabla
for _, row in candidates_data.iterrows():
    cursor.execute("""
    INSERT INTO candidates (First_Name, Last_Name, Email, Country, Application_Date, YOE, Seniority, Technology, Code_Challenge_Score, Technical_Interview_Score)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['First Name'], 
        row['Last Name'], 
        row['Email'], 
        row['Country'], 
        row['Application Date'], 
        row['YOE'], 
        row['Seniority'], 
        row['Technology'], 
        row['Code Challenge Score'], 
        row['Technical Interview Score']
    ))

# Confirmar los cambios
db_connection.commit()

# Cerrar la conexión
cursor.close()
db_connection.close()
```

Nueva Columna de Contratados ** No Tocar ni Ejecutar Mas ** 

```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine

try:
    # Establecer la conexión
    connection = mysql.connector.connect(**config)
    print("Conexión exitosa a la base de datos")

    # Crear un cursor para ejecutar consultas
    cursor = connection.cursor()

    # Ejecutar la consulta para obtener datos de la tabla original
    query = "SELECT * FROM candidates"
    cursor.execute(query)

    # Obtener los datos en un DataFrame de pandas
    df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    # Eliminar la columna 'id' si existe
    if 'id' in df.columns:
        df = df.drop(columns=['id'])

    # Agregar la columna 'contratado' basada en las condiciones
    df['contratado'] = (df['Code_Challenge_Score'] >= 7) & (df['Technical_Interview_Score'] >= 7)
    df['contratado'] = df['contratado'].map({True: 'Sí', False: 'No'})

    # Crear una nueva conexión usando SQLAlchemy para insertar el DataFrame en MySQL
    engine = create_engine(f'mysql+mysqlconnector://{config["user"]}:{config["password"]}@{config["host"]}/{config["database"]}')

    # Insertar el DataFrame en la tabla `candidates_2`, sin la columna 'id'
    df.to_sql(name='candidates_2', con=engine, if_exists='append', index=False)

    print("Los datos han sido insertados en la tabla 'candidates_2' sin la columna 'id'.")

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("Conexión cerrada")
```
